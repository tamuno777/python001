```python
import random 
numbers = []
counter = 1
while counter <= 50:
    num = random.randint(45,60)
    numbers.append(num)
    counter +=1
print(numbers)
```

```python
import math
sqrtnum = list(map(lambda x: math.sqrt(x),numbers))
print(sqrtnum)
```

```python
sqrtnum2 = list(filter(lambda x:x>45,numbers))
print(sqrtnum2)
```

```python
import random
numbers = []
counter = 1
while counter <= 100:
    num = random.randint(55,80)
    numbers.append(num)
    counter +=1
print(numbers)
```

```python
import math
sinnum = list(map(lambda x: math.sin(x),numbers))
print(sinnum)
```

```python
sinnum2 = list(filter(lambda x:x%2==0,numbers))
print(sinnum2)
```

```python
sinnum2 = list(filter(lambda x:x%3==0 and x%2!=0,numbers))
print(sinnum2)
```
