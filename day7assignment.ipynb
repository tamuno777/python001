```python
junior,intermediate,senior = 'JNR','ITM','SRN' 
Level = junior,intermediate,senior
print(Level)
```

```python
import random
ID = []
counter = 1
while counter <= 5000:
    num = random.randint(1,2000) 
    ID.append(num)
    counter +=1
print(ID)
```

```python
L = random.choice(Level)
sn = list(map(lambda L:f'{random.choice(Level)}{random.choice(ID)}',ID))
print(sn)
```

```python
sl = []
sl.append(sn)
print(sl)
```

```python
sl2 = list(filter(lambda sn:sn.startswith('JRN'),sl))
print(sl2)
```

```python
import random
ID = []
counter = 1
while counter <= 5000:
    num = random.randint(1,2000) 
    strnum = str(num)
    ID.append(junior + strnum)
    ID.append(intermediate + strnum)
    ID.append(senior + strnum)
    counter +=1
print(ID)
```

```python
sl2 = list(filter(lambda sn:sn.startswith('JNR'),ID))
print(sl2)
```
